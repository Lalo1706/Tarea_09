```python
class StackADT:
    def __init__(self):
        self.__data = list()

    def is_empty(self): #Regresa TRUE si la pila esta vacia.
         return len(self.__data) == 0
    
    def get_length(self):
        return len(self.__data)

    def pop(self):
        if self.is_empty():
            print("Pila vacia, nada que sacar")
        else:
            return self.__data.pop()
    
    def peek(self):
        if self.is_empty():
            print("Pila vacia, nada que sacar")
        else:
            return self.__data[-1]

    def push(self, value):
        self.__data.append(value)

    def to_string(self):
        for item in self.__data:
            print(f"| {item} |")
```

```python
entrada= open('code.txt', 'rt')
codigo = entrada.read()
pilaControl=StackADT()
pilaControl.push("#") 

for letra in codigo:
    if letra == '{':
        pilaControl.push("{")
    elif letra == '}':
        basura = pilaControl.pop()
    else:
        pass

for letra in codigo:
    if letra == '(':
        pilaControl.push("(")
    elif letra == ')':
        basura = pilaControl.pop()
    else:
        pass
if pilaControl.peek() == '#' :
    print("El código proporcionado esta correctamente balanceado.")
else :
    print("El código no esta balanceado en paréntesis ( y ) o en llaves { y }.")
```
